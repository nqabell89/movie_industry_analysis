```python
imdb_df = pd.read_csv('zippedData/imdb.title.principals.csv.gz')
imdb_df.head()
```

```python
budgets_df = pd.read_csv('zippedData/tn.movie_budgets.csv.gz')
budgets_df.head()
```

```python
genre_runtime_df = pd.read_csv('zippedData/imdb.title.basics.csv.gz')
genre_runtime_df.head()
```

```python
studio_df = pd.read_csv('zippedData/bom.movie_gross.csv.gz')
studio_df.head()
```

```python
rt_ratings_df = pd.read_csv('zippedData/rt.reviews.tsv.gz', sep='\t', encoding='unicode_escape')
rt_ratings_df.head()
```

```python
imdb_ratings_df = pd.read_csv('zippedData/imdb.title.ratings.csv.gz')
imdb_ratings_df.rename(columns={'averagerating':'avg_imdb_rating', 'numvotes':'rating_count'}, inplace=True)
imdb_ratings_df.head()
```

```python
budgets_df.info()
```

```python
budgets_df['title'] = budgets_df['movie']
budgets_df.drop(columns='movie',inplace=True)
budgets_df['title'].value_counts()
```

```python
budgets_df.head(30)
```

```python
genre_runtime_df['title'] = genre_runtime_df['primary_title']
genre_runtime_df.drop(columns='primary_title', inplace=True)
genre_runtime_df.head()
```

```python
test_df = pd.merge(budgets_df, genre_runtime_df, left_on='tconst', right_on='tconst', how='left')
test_df.head(20)
```

```python
test_df.info()
```

```python
working_df = pd.merge(imdb_ratings_df, test_df, left_on='tconst', right_on='tconst', how='inner')
working_df.info()
```

```python
working_nonull_df = working_df.dropna(how='any',axis=0)
df = working_nonull_df
```

```python
df['genres'] = df['genres'].map(lambda x: str(x).split(','))
```

```python
df['genres'].head()
```

```python
def remove_dollar_sign_commas(item):
    value = str(item)
    return value[2:]

test_item = '$ 4500600'
remove_dollar_sign_commas(test_item)
```

```python
temp_df = df.copy()
temp_df['budget'] = temp_df['production_budget'].map(remove_dollar_sign_commas)
temp_df['domestic'] = temp_df['domestic_gross'].map(remove_dollar_sign_commas)
temp_df['worldwide'] = temp_df['worldwide_gross'].map(remove_dollar_sign_commas)
temp_df.head()
```

```python
df = temp_df.copy()
df.budget = df.budget.astype(int)
df.domestic = df.domestic.astype(int)
df.info()
```

```python
sorted_df = df.sort_values('domestic')
sorted_df.drop(columns=['original_title', 'start_year', 'production_budget', 'domestic_gross', 'worldwide_gross'], inplace=True)
sorted_df.head(50)
```

```python
sorted_df['worldwide'] = sorted_df['worldwide'].astype(str)
sorted_df.drop(sorted_df[sorted_df['worldwide'] == '0' ].index , inplace=True)
sorted_df.head()
```

```python
sorted_df = sorted_df.sort_values(by='domestic',ascending=False)
sorted_df.head(30)
```
